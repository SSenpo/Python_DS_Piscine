```python
import pandas as pd
import sqlite3 as db
```

```python
connect = db.connect('../data/checking-logs.sqlite.sqlite')
cursor = connect.cursor()
```

```python
sql_reques = """
SELECT
	timestamp as datetime
FROM
	checker
WHERE
	uid LIKE 'user_%'
"""

commits_datatimes_df = pd.read_sql(sql=sql_reques, con=connect, parse_dates='datetime')
commits_datatimes_df.head()
```

```python
commits_datatimes_df.info()
```

```python
commits_datatimes_df['daytime'] = pd.cut(commits_datatimes_df['datetime'].dt.hour, 
										bins=[-1, 4, 10, 17, 24],
										labels=['night', 'morning', 'afternoon', 'evening'])
commits_datatimes_df
```

```python
daytime_counts_per_day_df = commits_datatimes_df.groupby(by=[commits_datatimes_df['datetime'].dt.date, commits_datatimes_df['daytime']]).count()
daytime_counts_per_day_df.rename(columns={'datetime': 'count'}, inplace=True)
daytime_counts_per_day_df = daytime_counts_per_day_df.unstack().droplevel(0, axis='columns')
daytime_counts_per_day_df
```

## Create graph

```python
plt = daytime_counts_per_day_df.plot.bar(figsize=(15, 8),
                                      fontsize=8,
                                      title='Commits per day',
                                      rot=90,
                                      alpha=0.4,
                                      stacked=True,
									  color=('black', 'r', 'blue', 'c')
                                     )
plt.grid(axis='y')
plt
```
